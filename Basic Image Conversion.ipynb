# Video Capturing

Video is nothing but a collection of frames which are displayed at a speed  termed as Frames Per Second(FPS).Here every frame is an image.

Here OpenCV processes the videos in a similar way it processes images. In OpenCV we can see video can be loaded from file or from camera device directly which made it more preferable choice for navigation systems like robots, embedded systems, automated vehicles and drone.

Here we are using cv2.VideoCapture() function in OpenCV to get video from file  or camera.

```python
#read a video from file
import numpy as np
import cv2
capture = cv2.VideoCapture("test.mp4")
# display the read video file
while capture.isOpened():
    ret, frame = capture.read()
    if not ret:
        break
    cv2.imshow("Video_window",frame)
    cv2.waitKey(25)
capture.release()
cv2.destroyAllWindows()
```

When Writing a video into files we need to know some properties like frame height, frame width, frame rate(fps).The following code helps in learning about the input video properties.

```python
capture = cv2.VideoCapture("test.mp4")
# get frame properties
print(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
print(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(capture.get(cv2.CAP_PROP_FRAME_COUNT))
```

To read video from device camera we need to replace file name in cv2.VideoCapture() with number 0 (zero).If this is the scenario, the user has to enable keyboard shortcut which helps to stop capturing. Below we have an example to show camera capturing along interrupting capturing using keyboard shortcut of letter ‘q’.

```python
import numpy as np
import cv2
capture = cv2.VideoCapture(0)
# display the read video file
while capture.isOpened():
    ret, frame = capture.read()
    if not ret:
        break
    cv2.imshow("Video_window",frame)
    cv2.waitKey(25) & 0xFF == ord(‘q’):
           break
capture.release()
cv2.destroyAllWindows()
```
